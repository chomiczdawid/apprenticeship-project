Kolejnym krokiem było wstępne przetworzenie danych, aby ułatwić proces translacji. W tym celu stworzona została funkcja *clean_text_round1*, której zadaniem było sprowadzić wszystkie słowa do małych liter, usunąć znaki interpunkcyjne, słowa w nawiasach, słowa zawierające liczby, znaki specjalne i entery. Tak przygotowane dane zostały zapisane do pliku csv i wprowadzone do arkusza Google Sheets w celu przetłumaczenia ich na język angielski. Do tego wykorzystano funkcje GOOGLETRANSLATE().

```python
import re
import string

def clean_text_round1(text):
    '''
    Małe litery, usunięcie interpunkcji, słów w nawiasach, słów zawierających liczby, innych znaków specjalnych, enterów
    '''
    text = text.lower()
    text = re.sub('\[.*?\;:]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('\n', '', text)
    return text

round1 = lambda x: clean_text_round1(x)
df_clean = pd.DataFrame(df.text.apply(round1))
```

```python
pd.set_option('max_rows',150)

df_clean.head(150)
```
